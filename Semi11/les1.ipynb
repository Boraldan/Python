Дана функция f(x) = 5x^2 + 10x - 30

1. Определить корни ++

2. Найти интервалы, на которых функция возрастает

3. Найти интервалы, на которых функция убывает

4. Построить график

5. Вычислить вершину

6. Определить промежутки, на котором f > 0

7. Определить промежутки, на котором f < 0

```python
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
```

```python
fx = 5*x**2 + 10*x - 30
fx
```

```python
plot(fx)
```

```python
solve(fx, x)
```

```python
x1, x2 = solve(fx)
round(x1.evalf(), 3), round(x2.evalf(), 3)
```

```python
solve(fx>0)
```

```python
solve(fx<0)
```

```python
diff(fx)
```

```python
plot(diff(fx))
```

```python
solve(diff(fx))
```

```python
solve(diff(fx)<0)
```

```python
solve(diff(fx)>0)
```
